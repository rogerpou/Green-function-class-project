```python
from ipywidgets import interact
import ipywidgets as widgets
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import scipy.integrate as integrate
```

```python
def green_fun(m,λ,T,A,f,nombre):
    ω=2*np.pi/T
    γ=λ/(2*m)
    Δt=T/100
    t=np.arange(0,10,Δt)
    G = lambda t: (1/(m*ω))*np.sin(ω*t)*np.exp(-γ*t)
    if nombre == 'dampeante':
        F = lambda t: A*np.cos(2 * np.pi * f * t)
    if nombre == 'sawtooth':
        F = lambda t: A*signal.sawtooth(2 * np.pi * f * t)
    if nombre == 'cuadrada':
        F = lambda t: A*signal.square(2 * np.pi * f * t)
    plt.plot(t,G(t))
    plt.plot(t,F(t))
    plt.title('Fuerza '+ nombre + ' a '+ str(f) +'Hz')
    plt.xlabel('Tiempo')
    plt.ylabel('Amplitud')
    plt.axhline(y=0, color='k')
    plt.grid(True, which='both')
    plt.xlim(0,10)
    val = integrate.trapz(G(t)*F(t),t)
    plt.text(2,.025,val)
    #return (1/(m*ω))*np.sin(ω*t)*np.exp(-γ*t)
interact(green_fun,m=(1,30,0.01),λ=(1,30,0.01),T=(0.01,4,0.01), A=(0.0001,0.01,0.0001),f=(0.01,3,0.01),nombre=['dampeante','sawtooth','cuadrada'])
   
```
